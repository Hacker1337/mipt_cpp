```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams.update({
    "text.usetex" : True,
    "font.family" : "serif",
    "font.serif" : "Computer Modern",
    "text.latex.preamble" : r'\usepackage{amsmath} \usepackage{mathtext} \usepackage[english, russian]{babel}'
})
plt.rcParams.update({
    "figure.figsize": (12, 7),
    'figure.dpi': 200
})
```

```python
def load_data(filename):
    a:np.array = np.loadtxt(filename)
    x:np.array = a[:, 0]
    ys:np.array = a[:, 1:]
    y = np.mean(ys, 1)
    dy = np.sqrt(np.mean(ys**2 - np.mean(ys, axis=1, keepdims=True)**2, axis = 1)/ys.shape[1])
    return x[1:], y[1:], dy[1:]
```

```python
x, y, dy = load_data("output10x10.txt")
# cd .. && make -f spacemake && cd space_fraction && ./spaceLauncher
# g++ map_launcher.cpp simulater.h simulater.cpp -o space_fraction/spaceLauncher
```
